```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
labels=['Dog','Cat','Horse']
```

```python
np.random.seed(1)
```

```python
W=np.random.randn(3,3072)
b=np.random.randn(3)
```

```python
test_image=cv2.imread(r"dog.4.jpg")
```

```python
img=cv2.resize(test_image,(32,32)).flatten()
```

```python
scores=np.dot(W,img)+b
```

```python
for (label,score)in zip(labels,scores):
    print(f"INFO {label}:{score}")
```

```python
cv2.putText(test_image,f"Label : {labels[np.argmax(scores)]}",(10,30),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
test_image=cv2.cvtColor(test_image,cv2.COLOR_BGR2RGB)
plt.imshow(test_image)
```
